```python
import pandas as pd
data = pd.read_csv('WHO_PM.csv')
```

```python
data.head()
data.info()
data.describe()
```

```python
data.dropna()
data.drop_duplicates()
```

```python
data.fillna('unknown', inplace=True)  # Replace NaN with 'unknown'
```

```python
data.drop(columns=['IndicatorCode', 'Indicator', 'ValueType', 'ParentLocationCode', 'Location type', 'SpatialDimValueCode', 'Period type', 'Dim1 type', 'Dim1ValueCode', 'Dim2 type', 'Dim2', 'Dim2ValueCode', 'Dim3 type', 'Dim3', 'Dim3ValueCode', 'DataSourceDimValueCode', 'DataSource', 'FactValueNumericPrefix', 'FactValueUoM', 'FactValueNumericLowPrefix', 'FactValueNumericHighPrefix', 'FactValueTranslationID', 'FactComments', 'Language', 'DateModified'], inplace=True)
```

```python
data.rename(columns={'ParentLocation': 'Region'}, inplace=True)
data.rename(columns={'Location': 'Country'}, inplace=True)
data.rename(columns={'Period': 'Year'}, inplace=True)
data.rename(columns={'Dim1': 'Residence Area Type'}, inplace=True)
```

```python
data.rename(columns={'FactValueNumeric': 'FineParticulateMatterCount'}, inplace=True)
data.rename(columns={'FactValueNumericLow': 'FineParticulateMatterLow'}, inplace=True)
data.rename(columns={'FactValueNumericHigh': 'FineParticulateMatterHigh'}, inplace=True)
```

```python
data.head()
data.info()
data.describe()
```

```python
data.to_csv('WHO_PM CLEANED.csv', index=False)
```
